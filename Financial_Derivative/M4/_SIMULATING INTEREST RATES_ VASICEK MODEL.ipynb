*In this lesson, we will tackle the use of stochastic differential equations (SDE) for simulating the behavior of other types of assets. Specifically, we will focus on simulating interest rates using the SDE associated with the **Vasicek (1977) model**.*

As always, let's first import some libraries we will need down the road:

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Vasicek (1977) Model

In Vasicek's model, the risk-neutral process for interest rates is:

$ dr = k(θ-r)dt + \sigma dz$

where $dz = \sqrt{dt} z$, with $z \sim \mathcal{N}(0,1)$. 

Let's build a function for this SDE:

```python
def vasicek(r0, K, theta, sigma, T, N, M):
    dt = T / N
    rates = np.zeros((N, M))
    rates[0, :] = r0
    for j in range(M):
        for i in range(1, N):
            dr = (
                K * (theta - rates[i - 1, j]) * dt
                + sigma * np.sqrt(dt) * np.random.normal()
            )
            rates[i, j] = rates[i - 1, j] + dr
    return rates
```

Now, we can use the same intuition of the Monte Carlo methods we used before for the Black-Scholes SDE to simulate interest rates given a current level, $r_0$. 

```python
M = 25  # Number of paths for MC
N = 150  # Number of steps
T = 1.0  # Maturity
r0 = 1.875
K = 0.25
theta = 0.02
sigma = 1.80
np.random.seed(42)  # For reproducibility
t = np.linspace(0, T, N)

rates = vasicek(r0, K, theta, sigma, T, N, M)
rates.shape
```

```python
import numpy as np

# Parameters for Question 19
S0 = 100    # Current stock price
mu = 0.10  # Annual return (10%)
sigma = 0.20 # Volatility (20%)
T = 1      # Time (1 year)
seed = 1

np.random.seed(seed)

# For one simulation, Z is a single random normal number
Z = np.random.normal(0, 1)

# Calculate ST for one simulation
ST = S0 * np.exp((mu - 0.5 * sigma**2) * T + sigma * np.sqrt(T) * Z)

# Calculate the change in price for this one simulation
change = (ST - S0)/S0

print(f'{change:.2f}')
```

```python
# average interest rate at T = 1 year across Monte Carlo paths
avg_r_T = rates[-1, :].mean()
std_r_T = rates[-1, :].std(ddof=1)
se_r_T = std_r_T / np.sqrt(M)
ci_lower, ci_upper = avg_r_T - 1.96 * se_r_T, avg_r_T + 1.96 * se_r_T

print(f"Average rate at T={T} year: {avg_r_T:.4f}")
print(f"Std dev across paths: {std_r_T:.4f}")
print(f"95% CI for the mean: ({ci_lower:.4f}, {ci_upper:.4f})")
```

We have just simulated interest rates for 100 paths and 100 steps.

Let's see how these look:

```python
for j in range(M):
    plt.plot(t, rates[:, j])

plt.xlabel("Time $t$", fontsize=14)
plt.ylabel("$r(t)$", fontsize=14)
plt.title("Vasicek Paths", fontsize=14)
axes = plt.gca()
axes.set_xlim([0, T])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```
