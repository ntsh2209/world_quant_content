```python
import matplotlib.pyplot as plt

fig = plt.figure()
np.random.seed(1)
T = 1
N = 255
dt = T / N
t = np.linspace(0, T, N)
M = 1
vol = 0.2
S0 = 100
mu = 0.1

dS = S0*(mu*dt + vol*np.sqrt(dt)*np.random.randn(N))
S = S0 + np.cumsum(dS,axis=1)

for i in range(M):
    plt.plot(t, S[i, :])

plt.xlabel('Time $t$',fontsize=14 )
plt.ylabel('Random VariableStock Price $S_t$',fontsize=14)
plt.title('Simulated Stock Price Paths',fontsize=16)
axes = plt.gca()
axes.set_xlim([0, T])
plt.xticks([0,T])
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.tight_layout()
plt.show()
```
