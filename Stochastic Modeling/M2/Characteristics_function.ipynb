As we have already seen, one important input in Fourier-type option pricing methods is the characteristic function (CF) of the underlying stock price process. Here we dig a little bit deeper into the definition of CF.

## **1. Characteristic Function (CF)**

The CF will come in handy because, for complex methods, directly observing the PDF is impossible and/or too computationally costly.

In this lesson, we will go over the mathematical foundations of CF and see a step-by-step example on how to derive it for a simple (Black-Scholes type) stock price process.

The **characteristic function**, $\varphi_X(u)$, of a real valued random variable $X$ depicts the **distribution function uniquely**. It is defined as:

\
$$
\varphi_X (u) = \mathbb{E}[e^{iuX}]\ , \ with \ i = \sqrt{-1}
$$

\
Which we can onveniently rewrite as:

\
$$
\varphi_X (u) = \int_{-\infty}^{\infty} e^{iuX} \,dF(x) = \int_{-\infty}^{\infty} e^{iuX}f(x) \,dx
$$

\
where $F(x)$ and $f(x)$ denote the CDF and PDF of $X$, respectively.

## **2. Deriving CF: Black-Scholes example**

Next, we will go over the process of deriving a specific characteristic function from scratch. We will do that for the **Black-Scholes** model. However, the process is common to other models.

Before doing that, let's define a **Lemma** that will help us later on in the derivation.

- **Lemma**. Define $y = \hat{\mu}+i u \hat{\sigma}^2$, then:

\
$$
ius_T - \frac{(s_T - \hat{\mu})^2}{2\hat{\sigma}^2} = \frac{-(s_T - y)^2 + 2\hat{\mu}\hat{\sigma}^2 iu - u^2 \hat{\sigma}^4}{2\hat{\sigma}^2}
$$

\
**Proof**

\
$$
ius_T - \frac{(s_T - \hat{\mu})^2}{2\hat{\sigma}^2} = \frac{2\hat{\sigma}^2 ius_T - (s_T^2 -2s_T\hat{\mu}+ \hat{\mu}^2) }{2\hat{\sigma}^2} = \frac{-s_T^2 + 2s_T\hat{\mu} + 2s_T iu\hat{\sigma}^2 - \hat{\mu}^2}{2\hat{\sigma}^2} = \frac{-(s_T - y)^2 + 2\hat{\mu}\hat{\sigma}^2 iu - u^2 \hat{\sigma}^4}{2\hat{\sigma}^2}
$$


### 2.1. The Black-Scholes CF

Now, we will work on deriving the CF of a log-normal process, as is the case with stock prices in Black-Scholes.

Let's start by working with a normally-distributed variable $s = Ln S$, with $\mathbb{Q}(s_T)$ its risk-neutral PDF:

\
$$
\varphi_{s_T} (u) = \mathbb{E}[e^{ius_T}] = \int_{-\infty}^{\infty} e^{ius_T} \mathbb{Q}(s_T) \,ds_T = \int_{-\infty}^{\infty} e^{ius_T} \frac{1}{\sigma \sqrt{2\pi T}} e^{\frac{-\{s_T-[s_0 + (\mu - \frac{\sigma^2}{2})T]\}^2}{2\sigma^2T}}  \,ds_T = \frac{1}{\sigma \sqrt{2\pi T}} \int_{-\infty}^{\infty} e^{ius_T - \frac{\{s_T-[s_0 + (\mu - \frac{\sigma^2}{2})T]\}^2}{2\sigma^2T}} \,ds_T
$$


\
Let $\hat{\mu}= \left[ s_0 + \left( \mu - \frac{\sigma^2}{2}\right) T \right]$ and $\hat{\sigma}= \sigma \sqrt{T}$, then:


\
$$
\varphi_{s_T} (u) = \frac{1}{\sigma \sqrt{2\pi T}} \int_{-\infty}^{\infty} e^{ius_T - \frac{\{s_T-[s_0 + (\mu - \frac{\sigma^2}{2})T]\}^2}{2\sigma^2T}} \,ds_T = \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{ius_T - \frac{(s_T - \hat{\mu})^2 }{2\hat{\sigma}^2}} \,ds_T
$$

\
Now, the **Lemma** defined before comes in handy. Applying it:

\
$$
\varphi_{s_T} (u) = \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{ius_T - \frac{(s_T - \hat{\mu})^2 }{2\hat{\sigma}^2}} \,ds_T = \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{\frac{-(s_T - y)^2 + 2\hat{\mu}\hat{\sigma}^2 iu - u^2 \hat{\sigma}^4}{2\hat{\sigma}^2}} \,ds_T = e^{\frac{2\hat{\mu}\hat{\sigma}^2iu-u^2\hat{\sigma}^4}{2\hat{\sigma}^2}}\frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{\frac{-(s_T - y)^2}{2\hat{\sigma}^2}} \,ds_T = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2} \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-\left( \frac{s_T - y}{\sqrt{2\hat{\sigma}^2}} \right) ^2} \,ds_T
$$


\
Next, consider $z = \frac{s_T-y}{\sqrt{2\hat{\sigma}^2}}$, and the closed upper and lower bounds of $s_T$, $[-\gamma, \gamma]$,

\
$$
z = \frac{s_T - (\hat{\mu}+ iu\hat{\sigma}^2)}{\sqrt{2\hat{\sigma}^2}} = \frac{s_T - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}} - \frac{iu\hat{\sigma}}{\sqrt{2}} \\
     \varphi_{s_T} (u) = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2} \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{\frac{-\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}}^{\frac{\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}} e^{-z^2}\sqrt{2\hat{\sigma}^2} \,dz
$$

At this point, **Cauchy's integral theorem** will help in solving the integral analtically.

- **Cauchy's integral theorem**. If a function $f(z)$ is analytic in a simply connected domain $\Omega$, then for any simply closed contour $C$ in $\Omega$,

\
$$
    \int_C f(z) \,dz = 0
$$

\
Let $a_i$, for $i=1,2,3,4$ denote the corners of the rectangular contour $C$:

\
$$
a_1 = \frac{\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}, \ \ a_2 = \frac{-\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}, \ \ a_3 = \frac{-\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}, \ \ a_4 = \frac{\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}
$$

\
By Cauchy's theorem we know that,

\
$$
\oint_{C \in \Omega} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz = 0
\Leftrightarrow \int_{a_1}^{a_2} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz + \int_{a_2}^{a_3} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz + \int_{a_3}^{a_4} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz + \int_{a_1}^{a_4} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz = 0
$$

\
We can now solve the previous integrals separately and check their behavior when $\gamma \rightarrow \infty$. Let's start with those that do not contain any imaginary parts in the bounds:

\
$$
    \lim_{\gamma \rightarrow \infty} \int_{a_1}^{a_2} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz  = -\lim_{\gamma \rightarrow \infty} \sqrt{2\hat{\sigma}^2} \int_{a_2}^{a_1} e^{-z^2} \,dz = - \sqrt{2\pi \hat{\sigma}^2}
$$


Since we know that $\int_{-\infty}^{\infty} e^{-z^2} \,dz = \sqrt{\pi}$. Next, let's check those integrals with imaginary components in only one of the bounds, for example that between $a_4$ and $a_1$:

\
$$
    \lim_{\gamma \rightarrow \infty} \int_{a_4}^{a_1} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz  = \lim_{\gamma \rightarrow \infty} \sqrt{2\hat{\sigma}^2} \int_{a_4}^{a_1} e^{-z^2} \,dz = 0
$$

\
It is easy to check that these will be $0$, since we know that, for example, $\int_{\infty-iz}^{\infty} e^{-z} \,dz = 0$.

\
Finally, to fulfill Cauchy's theorem we know that:

\
$$
\int_{a_3}^{a_4} \sqrt{2\hat{\sigma}^2} e^{-z^2} \,dz  = \sqrt{2\pi \hat{\sigma}^2}
$$

\
Plugging the previous result of the characteristic function expression we had from before:

\
$$
\varphi_{s_T} (u) = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2} \frac{1}{\hat{\sigma} \sqrt{2\pi}} \int_{\frac{-\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}}^{\frac{\gamma - \hat{\mu}}{\sqrt{2\hat{\sigma}^2}}-\frac{iu\hat{\sigma}}{\sqrt{2}}} e^{-z^2}\sqrt{2\hat{\sigma}^2} \,dz = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2} \frac{\sqrt{2\pi \hat{\sigma}^2}}{\hat{\sigma} \sqrt{2\pi}} = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2}
$$

\
Replacing back $\hat{\mu}= \left[ s_0 + \left( \mu - \frac{\sigma^2}{2}\right) T \right]$ and $\hat{\sigma}= \sigma \sqrt{T}$, yields the **Black-Scholes CF**

\
$$
\varphi_{s_T} (u) = e^{\hat{\mu}iu - \frac{1}{2}u^2\hat{\sigma}^2} = e^{\left(s_0 + \left(\mu - \frac{\sigma^2}{2}\right) T\right)iu - \frac{1}{2}u^2 \sigma^2 T}
$$
